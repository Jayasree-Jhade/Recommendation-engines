```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv("movies.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df["genres"]=df["genres"].fillna("")
```

```python
df.isnull().sum()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer=TfidfVectorizer()
feature_vector=vectorizer.fit_transform(df["genres"])
print(feature_vector)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity=cosine_similarity(feature_vector)
print(similarity)
```

```python
print(similarity.shape)
```

```python
# create a list with all movie names given in dataset
```

```python
list1=df["title"].tolist()
```

```python
print(list1)
```

```python
# getting the movie name from user
```

```python
movie_name=input("enter the movie name")
import difflib
match=difflib.get_close_matches(movie_name,list1)
print(match)
```

```python
close_match=match[0]
print(close_match)
```

```python
index=df[df.title==close_match]["index"].values[0]
print(index)
```

```python
similarity_score=list(enumerate(similarity[index]))
print(similarity_score)
```

```python
len(similarity_score)
```

```python
sorted_similar_movies=sorted(similarity_score,key=lambda x:x[1],reverse=True)
print(sorted_similar_movies)
```

```python
print("movies suggested for u")
i=1
for movie in sorted_similar_movies:
    index=movie[0]
    title_from_index=df[df.index==index]["title"].values[0]
    if(i<=10):
        print(i,  ".",title_from_index)
    i=i+1
```
